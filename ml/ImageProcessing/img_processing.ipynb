## Affine transformation

```python
img9 = cv2.imread('glass.jpg')
rows, cols, _ = img.shape
```

```python
plt.imshow(img9)
```

```python
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
```

```python
M = cv2.getAffineTransform(pts1, pts2)
new6 = cv2.warpAffine(img9, M, (cols,rows))
plt.imshow(new6)
plt.show()
```

## Affine transformation on another image

```python
img10 = cv2.imread('planet.jpg')
rows, cols, _ = img10.shape
plt.imshow(img10)
```

```python
pts1 = np.float32([[200,40],[550,40],[200,550]])
pts2 = np.float32([[400,40],[750,40],[400,750]])
M = cv2.getAffineTransform(pts1, pts2)
new6 = cv2.warpAffine(img10, M, (cols,rows))
plt.imshow(new6)
plt.show()
```

## Perspective Transformation

```python
pts1 = np.float32([[75,20],[145,20],[85,200],[140,200]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
M = cv2.getPerspectiveTransform(pts1, pts2)
new7 = cv2.warpPerspective(img9, M, (cols,rows))
plt.imshow(new7)
plt.show()
```

```python
M = np.float32([[1,0,80],[0,1,70]])
new8 = cv2.warpAffine(img9, M, (cols,rows))
new8.shape
plt.imshow(new8)
```

## Rotating the image

## Handling noise in text image

```python
img11 = cv2.imread('j.png')
plt.imshow(img11)
```

### Erode the image with 1 iteration

```python
kernel = np.ones((7,7), np.uint8)
erosion = cv2.erode(img11, kernel, iterations=1)
plt.imshow(erosion)
```

### kernel having a size...slides through the image from start till end. when erosion sees black color anywhere it replaces black color in center of kernel box

### Dilation

```python
dilation = cv2.dilate(img11, kernel, iterations=1)
plt.imshow(dilation)  #dilation puts white pixels
```
